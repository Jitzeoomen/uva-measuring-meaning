# Multidimensional scaling

Included in R is the dataset `eurodist` which contains the distances Between European Cities and Between US Cities.

```r
data(eurodist)
ls()
```

```r
# Should be 'dist'; a distance structure.
class(eurodist)
```

```r
# See what the structure looks like as a matrix.
# Can compare this to matrix 6.1
head(as.matrix(eurodist))
```

```r
# Classical multidimensional scaling (MDS) of a data matrix.
loc <- cmdscale(eurodist)
head(loc)
```

```r
x <- loc[, 1]
y <- -loc[, 2] # reflect so North is at the top
```

```r
plot(x, y, 
     type = "n", 
     xlab = "", 
     ylab = "", 
     asp = 1, 
     axes = FALSE,
     main = "Cmdscale(eurodist)")
text(x, y, rownames(loc), cex = 0.8)
```

```r
# Nonmetric MDS

n_fit <- MASS::isoMDS(eurodist, k = 2)
```

```r
n_fit
```

```r
n_x <- n_fit$points[, 1]
n_y <- -n_fit$points[, 2]
```

```r
plot(n_x, n_y, 
     xlab="", 
     ylab="", 
     main="Nonmetric MDS", 
     axes = FALSE,
     type="n")
text(n_x, n_y, labels = row.names(as.matrix(eurodist)), cex = .7)
```
